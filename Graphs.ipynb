```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

## INSERT

```python
data_insert = pd.read_csv('insert.csv', sep=';').to_dict()
```

```python
number_of_elements = list(data_insert['AVL'].keys())
AVL_insert_time = list(data_insert['AVL'].values())
Splay_insert_time = list(data_insert['Splay'].values())
Cartesian_insert_time = list(data_insert['Cartesian'].values())
RedBlack_insert_time = list(data_insert['Red-Black'].values())
lg = [math.log(n) for n in number_of_elements]
```

```python
plt.plot(number_of_elements, AVL_insert_time, label = "AVL")
plt.plot(number_of_elements, Splay_insert_time, label = "Splay")
plt.plot(number_of_elements, Cartesian_insert_time, label = "Cartesian")
plt.plot(number_of_elements, RedBlack_insert_time, label = "RedBlack")
plt.plot(number_of_elements, lg, label="LOG")
plt.legend()
```
