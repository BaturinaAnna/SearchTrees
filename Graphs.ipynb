## The graphs show the dependence of the execution time of one operation (in milliseconds) on the size of the binary search tree 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

## INSERT

```python
data_insert = pd.read_csv('insert.csv', sep=';').to_dict()
```

```python
number_of_elements = list(data_insert['Number of elements'].values())
AVL_insert_time = list(data_insert['AVL'][i]/10000 for i in range(len(data_insert['AVL'].values())))
Splay_insert_time = list(data_insert['Splay'][i]/10000 for i in range(len(data_insert['AVL'].values())))
Cartesian_insert_time = list(data_insert['Cartesian'][i]/10000 for i in range(len(data_insert['AVL'].values())))
RedBlack_insert_time = list(data_insert['Red-Black'][i]/10000 for i in range(len(data_insert['AVL'].values())))
```

```python
plt.plot(number_of_elements, AVL_insert_time, label = "AVL")
plt.plot(number_of_elements, Splay_insert_time, label = "Splay")
plt.plot(number_of_elements, Cartesian_insert_time, label = "Cartesian")
plt.plot(number_of_elements, RedBlack_insert_time, label = "RedBlack")
plt.legend()
```
